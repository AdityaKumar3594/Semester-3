*** Filter based methods***
    
     SelectKBest: It selects k highest scoring features based on a function and removes the rest of the features.
     Lets take an example of California Housing dataset.
     
     import numpy as np
     from sklearn.datasets import fetch_california_housing
     …………
1)download data and select a subset X_california[:2000, :], y_california[:2000]

2)Calculate Shape of feature matrix before feature selection

3)Select 3 most important features, since it is a regression problem, we can use only mutual_info_regression or f_regression scoring functions only. Calculate shape of feature matrix after feature selection






```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_regression,f_regression
```

```python
X_california = fetch_california_housing().data
y_california = fetch_california_housing().target
```

```python
X_subset = X_california[:2000, :]
y_subset = y_california[:2000]
```

```python
print(f"Shape of feature matrix before feature selection: {X_subset.shape}")
X_subset
```

```python
selector = SelectKBest(score_func=mutual_info_regression, k=3)
X_new = selector.fit_transform(X_subset, y_subset)

# Calculate the shape of the feature matrix after feature selection
print(f"Shape of feature matrix after feature selection: {X_new.shape}")
X_new

```

```python
f_selector = SelectKBest(score_func=f_regression, k=3)
X_new = f_selector.fit_transform(X_subset, y_subset)

# Calculate the shape of the feature matrix after feature selection
print(f"Shape of feature matrix after feature selection: {X_new.shape}")
X_new
```

    • SelectPercentile: This is very similar to SelectKBest from previous section, the only difference is, it selects top Percentile of
      all features and drops the rest of the features.
      Similar to SelectKBest, it also uses a scoring function to decide importance of features.Let's use the california housing price
      dataset also for this API and calculate top 30 percentile of features.

```python
from sklearn.feature_selection import SelectPercentile
```

```python
S_selector = SelectPercentile(score_func=mutual_info_regression, percentile=30)
X_new = S_selector.fit_transform(X_subset, y_subset)

# Calculate the shape of the feature matrix after feature selection
print(f"Shape of feature matrix after feature selection: {X_new.shape}")
X_new
```

    • Similarly apply GenericUnivariateSelect also for same number of features on the same dataset.


```python
from sklearn.feature_selection import GenericUnivariateSelect

GU_selector = GenericUnivariateSelect(score_func=mutual_info_regression, mode='k_best', param=3)
X_new = GU_selector.fit_transform(X_subset, y_subset)

# Calculate the shape of the feature matrix after feature selection
print(f"Shape of feature matrix after feature selection: {X_new.shape}")
X_new
```
